### 1. Yelp Dataset: business.json

####  - The JSON format is as following
```

{
    // string, 22 character unique string business id
    "business_id": "tnhfDv5Il8EaGSXZGiuQGg",

    // string, the business's name
    "name": "Garaje",

    // string, the full address of the business
    "address": "475 3rd St",

    // string, the city
    "city": "San Francisco",

    // string, 2 character state code, if applicable
    "state": "CA",

    // string, the postal code
    "postal code": "94107",

    // float, latitude
    "latitude": 37.7817529521,

    // float, longitude
    "longitude": -122.39612197,

    // float, star rating, rounded to half-stars
    "stars": 4.5,

    // integer, number of reviews
    "review_count": 1198,

    // integer, 0 or 1 for closed or open, respectively
    "is_open": 1,

    // object, business attributes to values. note: some attribute values might be objects
    "attributes": {
        "RestaurantsTakeOut": true,
        "BusinessParking": {
            "garage": false,
            "street": true,
            "validated": false,
            "lot": false,
            "valet": false
        },
    },

    // an array of strings of business categories
    "categories": [
        "Mexican",
        "Burgers",
        "Gastropubs"
    ],

    // an object of key day to value hours, hours are using a 24hr clock
    "hours": {
        "Monday": "10:00-21:00",
        "Tuesday": "10:00-21:00",
        "Friday": "10:00-21:00",
        "Wednesday": "10:00-21:00",
        "Thursday": "10:00-21:00",
        "Sunday": "11:00-18:00",
        "Saturday": "10:00-21:00"
    }
}

```




````python
business_df = pd.read_csv('csv_out/business.csv',low_memory=False)
    
business_df.head()
    
````

````python
business_df.count()
````

````python
business_ON_df = business_df[(business_df['state']=='ON')|(business_df['state']=='on')]
business_ON_df.count()
````

````python
business_ON_df.head()
````

### 2. Yelp Dataset: checkin.json

####  - The JSON format is as following

Checkins on a business.

{
    // string, 22 character business id, maps to business in business.json
    "business_id": "tnhfDv5Il8EaGSXZGiuQGg"

    // string which is a comma-separated list of timestamps for each checkin, each with format YYYY-MM-DD HH:MM:SS
    "date": "2016-04-26 19:49:16, 2016-08-30 18:36:57, 2016-10-15 02:45:18, 2016-11-18 01:54:50, 2017-04-20 18:39:06, 2017-05-03 17:58:02"
}


````python
checkin_df = pd.read_csv('csv_out/checkin.csv')

checkin_df.head()
````

````python
checkin_df.count()
````

````python
checkin_df.columns = ['business_id','checkin_date']
checkin_df.head()
````

````python
business_ON_checkin_df = pd.merge(business_ON_df,checkin_df,on='business_id',how='left')
business_ON_checkin_df.head()
````

````python
business_ON_checkin_df.count()
````

### 3. Yelp Dataset: tip.json

#### - The JSON format is as following

Tips written by a user on a business. Tips are shorter than reviews and tend to convey quick suggestions.

{
    // string, text of the tip
    "text": "Secret menu - fried chicken sando is da bombbbbbb Their zapatos are good too.",

    // string, when the tip was written, formatted like YYYY-MM-DD
    "date": "2013-09-20",

    // integer, how many compliments it has
    "compliment_count": 172,

    // string, 22 character business id, maps to business in business.json
    "business_id": "tnhfDv5Il8EaGSXZGiuQGg",

    // string, 22 character unique user id, maps to the user in user.json
    "user_id": "49JhAJh8vSQ-vM4Aourl0g"
}

````python
tip_df = pd.read_csv('csv_out/tip.csv')
tip_df.head()
````

````python
tip_df.count()
````

````python
tip_df.columns = ['business_id','compliment_count','tip_date','tip_text','user_id']
tip_df.head()
````

````python
business_ON_checkin_tip_df = pd.merge(business_ON_checkin_df,tip_df,on='business_id',how='left')
business_ON_checkin_tip_df.head()
````

````python
business_ON_checkin_tip_df.count
````
